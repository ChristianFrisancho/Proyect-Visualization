```python
import pandas as pd
import numpy as np
from scipy import stats
year_cols = [c for c in df2.columns if c.startswith("F")]
Z = np.abs(stats.zscore(df2[year_cols], nan_policy='omit'))
TH = 3
mask_out = (Z > TH).any(axis=1)
outliers = df2.loc[mask_out].copy()
df_sin_outliers = df2.loc[~mask_out].copy()
outliers
```

```python
df3 = df2.loc[~mask_out].reset_index(drop=True).copy()
df3
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (9,5)
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False

year_cols = [c for c in df3.columns if c.startswith("F")]

display(df3[year_cols].describe().T.round(3))

YEAR = "F2021"
fig, ax = plt.subplots()
ax.hist(df3[YEAR].dropna(), bins=40)
ax.set_title(f"Distribución {YEAR}")
ax.set_xlabel(YEAR); ax.set_ylabel("frecuencia")
ax.set_yscale("log")  
plt.show()

if "Technology" in df3.columns:
    order = (
        df3.groupby("Technology")[YEAR]
           .median()
           .sort_values(ascending=False)
           .index
    )
    data = [df3.loc[df3["Technology"]==t, YEAR].dropna() for t in order]
    fig, ax = plt.subplots()
    ax.boxplot(data, labels=order, vert=True, showfliers=True)
    ax.set_title(f"{YEAR} por Tecnología")
    ax.set_ylabel("valor"); plt.xticks(rotation=30, ha="right")
    plt.show()

for col in ["Technology", "CTS_Name", "Unit"]:
    if col in df3.columns:
        ax = df3[col].value_counts().plot(kind="bar", rot=0)
        ax.set_title(f"Conteo por {col}"); ax.set_xlabel(""); ax.set_ylabel("count")
        plt.show()

global_ts = df3[year_cols].sum().rename_axis("year").reset_index(name="value")
global_ts["year"] = global_ts["year"].str[1:].astype(int)
ax = global_ts.plot(x="year", y="value", legend=False)
ax.set_title("Suma global por año"); ax.set_xlabel("Año"); ax.set_ylabel("valor")
plt.show()
```

```python
import numpy as np

corr = df3[year_cols].corr()
fig, ax = plt.subplots(figsize=(6,5))
im = ax.imshow(corr.values, cmap="viridis")
ax.set_title("Correlación entre años")
ax.set_xticks(range(len(year_cols))); ax.set_yticks(range(len(year_cols)))
ax.set_xticklabels([c[1:] for c in year_cols], rotation=90, fontsize=7)
ax.set_yticklabels([c[1:] for c in year_cols], fontsize=7)
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.show()

subset = [c for c in ["F2019","F2020","F2021","F2022","F2023"] if c in df3.columns]
pd.plotting.scatter_matrix(df3[subset].dropna(), figsize=(9,9), diagonal="hist", alpha=0.6, s=12)
plt.suptitle("Matriz de pares (años recientes)", y=1.02)
plt.show()

YEAR = "F2021"
if "Technology" in df3.columns and "Country" in df3.columns:
    piv = (
        df3.pivot_table(index="Country", columns="Technology", values=YEAR, aggfunc="sum")
           .rename(columns={"Hydropower excl. PS":"Hydro"})
           .fillna(0.0)
    )
    xcol, ycol = ("Solar", "Wind") 
    if xcol in piv.columns and ycol in piv.columns:
        fig, ax = plt.subplots()
        ax.scatter(piv[xcol]+1e-9, piv[ycol]+1e-9, s=25, alpha=0.8)
        ax.set_title(f"{xcol} vs {ycol} — {YEAR}")
        ax.set_xlabel(xcol); ax.set_ylabel(ycol)
        ax.set_xscale("log"); ax.set_yscale("log")
        plt.show()

if "Technology" in df3.columns:
    tech_ts = (
        df3.groupby("Technology")[year_cols]
           .sum()
           .T.rename_axis("year")
           .reset_index()
    )
    tech_ts["year"] = tech_ts["year"].str[1:].astype(int)
    vals = tech_ts.set_index("year")
    fig, ax = plt.subplots()
    ax.stackplot(vals.index, *[vals[c].values for c in vals.columns], labels=vals.columns)
    ax.set_title("Stacked area por Tecnología"); ax.set_xlabel("Año"); ax.set_ylabel("valor")
    ax.legend(ncol=2, fontsize=9, loc="upper left")
    plt.show()

if "Technology" in df3.columns and "Country" in df3.columns:
    has_val = df3[[ "Country","Technology", YEAR ]].dropna()
    cross = pd.crosstab(has_val["Country"], has_val["Technology"])
    display(cross.head())
```
